```python
import pandas as pd
```

Создаём ДФ для таблицы с рейтингом

```python
df_ratings = pd.read_csv("u.data.csv", names=['user_id', 'movie_id', 'rating', 'timestamp'], header=None, sep='\t', encoding = "ISO-8859-1")
df_ratings.head()
```

ДФ для таблицы с фильмами

```python
df_movies = pd.read_csv("u.item.csv", names=['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation',
              'Children',  'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'], header=None, sep='|', encoding = "ISO-8859-1")
df_movies.head()
```

3) Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь

```python
df_ratings = df_ratings.loc[df_ratings['user_id']==405]
df_ratings.head()
```

4) Добавьте к датафрейму из задания 3 столбцы: По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей

```python
count_ratimg = df_ratings['movie_id'].value_counts().reset_index().rename(columns={'index':'movie_id', 'movie_id':'count_ratings'})
count_ratimg.head()
```

```python
sum_rating = df_ratings[['movie_id','rating']].groupby('movie_id').sum().reset_index().rename(columns={'rating':'sum_ratings'})
sum_rating.head()
```

```python
df_movies['release_date']=pd.to_datetime(df_movies['release_date'])
df_movies['year'] = df_movies['release_date'].dt.year
df_movies.head()
```

Соединяем полученные результаты

```python
df_r1=df_movies.merge(sum_rating, left_on='movie_id', right_on='movie_id', how='inner')
df_r1.head()
```

```python
df_r1=df_r1.merge(count_ratimg, left_on='movie_id', right_on='movie_id', how='inner')
df_r1.head()
```

```python
df_r1=df_r1.merge(df_ratings, left_on='movie_id', right_on='movie_id', how='inner')
df_r1.head()
```

Удаляем лишние колонки

```python
df_r1 =df_r1.drop(['release_date','movie_id','movie_title','video_release_date','IMDb_URL','user_id','timestamp'],axis=1)
df_r1.head()
```

9) Средствами спарка вывести среднюю оценку для каждого фильма

```python
df_d=spark.read.csv('u.data.csv',header=None,sep='\t',inferSchema=True)
df_d.show()
```

```python
df_d=df_d.toDF('user_id','movie_id','rating','timestamp')
df_d.show()
```

```python
df_i=spark.read.csv('u.item.csv',header=None,sep='|',inferSchema=True)
df_i.show()
```

```python
df_i=df_i.toDF('movie_id','movie_title','release_date','video_release_date','IMDb_URL','unknown','Action','Adventure','Animation',
              'Children','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western')
df_i.show()
```

```python
df_j=df_i.join(df_d, 'movie_id', how= 'inner')
```

```python
df_j.show()
```

```python
df_avg= df_j.groupBy ('movie_title'). avg('rating')
df_avg.show()
```
